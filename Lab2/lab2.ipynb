```python
#Read the csv, notice sep=','.. For CSV files created in Norwegian excel we need to use sep=';'#
#df = pd.read_csv(url, na_values=' ') You can use this for automatically convert blank space to nan values
df = pd.read_csv(url, sep=',')

df.head()
```

```python
df.isna().sum()
```

```python
# Mean age
print('Mean of "Age" is %.2f' %(df["Age"].mean(skipna=True)))
# Median age
print('Median of "Age" is %.2f' %(df["Age"].median(skipna=True)))
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df.isna().sum()
```

```python
df["Fare"] = df["Fare"].fillna(df["Fare"].median())
```

```python
df["Embarked"] = df["Embarked"].fillna("S")
```

```python
df["HasCabin"] = df.Cabin.isnull()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['HasCabin'] = ~df.Cabin.isnull()
df.head()
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df["Title"] = df.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1)) 
```

```python
df.head()
```

```python
df["Title"].value_counts()
```

```python

# Change Ms to miss, Mme to Mr
df["Title"] = df["Title"].replace({'Mme':'Mrs', 'Ms':'Miss', 'Mlle':'Miss'})
# Unique category
df["Title"] = df["Title"].replace(["Rev","Dr", "Col", "Major", "Lady", "Sir", "Don", "Capt", "Countess", "Jonkheer", "Dona"], "Unique")
df["Title"].value_counts()
```

```python
# Create a new column that extracts the title from the "Name" column
df['Name_Title'] = df['Name'].apply(lambda name: re.search(' ([A-Za-z]+)\.', name).group(1))

# Find rows where "Name_Title" does not match "Title"
mismatched_titles = df[df['Name_Title'] != df['Title']]
```

```python
# Find the index of the rows where 'Mme' is in the 'Name' column
indices = df[df['Name'].str.contains('Mme')].index

# Change the 'Title' of these rows
df.loc[indices, 'Title'] = 'Mrs'
```

```python
df["Title"].value_counts()
```

```python
df = df.drop('Name_Title', axis=1)
```

```python
sns.countplot(x='Title', data=df)
plt.xticks(rotation=45)
```

```python
df['CatAge'] = pd.qcut(df["Age"], q=4, labels=False )
df['CatFare'] = pd.qcut(df["Fare"], q=4, labels=False )
```

```python
df.head()
```

```python
#Dropping columns that wont make sense in binary format
df = df.drop(["Age", "Fare", "Cabin", "Name", "Ticket"], axis=1)
df.head()
```

```python
# Converting to binary values
df_dum = pd.get_dummies(df, drop_first=True)
df_dum.head()
```

```python
df_dum.to_csv('data/Titanic_Cleaned.csv', index=False)
```
