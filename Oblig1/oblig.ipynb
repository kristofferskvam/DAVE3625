```python
#Import modules
%matplotlib inline
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Create DataFrame
Grc_df = pd.read_csv('data/Grocery_Dataset.csv')

# Task 1
# First 10 rows
Grc_df.head(10)
```

```python
# Task 1
# Last 10 rows
Grc_df.tail(10)
```

```python
# Task 2
# Number of null values in the column
Grc_df.isna().sum()
```

```python
print('Percent missing in "Item_Weight" column: %.2f%%' %((Grc_df['Item_Weight'].isnull().sum()/Grc_df.shape[0])*100))
print('Percent missing in "Outlet_Size" column: %.2f%%' %((Grc_df['Outlet_Size'].isnull().sum()/Grc_df.shape[0])*100))
```

```python
# Find median item weight
print('Median of "Item_Weight" is %.2f' %(Grc_df["Item_Weight"].median(skipna=True)))
```

```python
# Fill missing Item_Weight values with median weight.
Grc_df["Item_Weight"] = Grc_df["Item_Weight"].fillna(Grc_df["Item_Weight"].median())
```

```python
# Find most common value in Outlet_Size, completes both tasks 2 and 3.
# 3 Outlet sizes, small is minimum, high is maximum.
sns.countplot(x='Outlet_Size', data=Grc_df, hue='Outlet_Size')
plt.show()
```

```python
# We'll fill Outlet_Size with the most common value, which is "Medium".
Grc_df["Outlet_Size"] = Grc_df["Outlet_Size"].fillna("Medium")
```

```python
# Task 4. Lists item fat content
sns.countplot(x='Item_Fat_Content', data=Grc_df, hue='Item_Fat_Content')
plt.show()
```

```python
# Notice that there are 3 values that describe low fat, and 2 values that describe regular.
# Replaces the values to just one for each category and display the results
Grc_df['Item_Fat_Content'] = Grc_df['Item_Fat_Content'].replace({'low fat':'Low Fat', 'LF':'Low Fat', 'reg':'Regular'})
sns.countplot(x='Item_Fat_Content', data=Grc_df, hue='Item_Fat_Content')
plt.show()
```

```python
# Task 5
# Make Grc_new_df a copy of Grc_df. Display the columns
Grc_new_df = Grc_df
Grc_new_df.columns
```

```python
# Drop the columns having 0 and 6 as indexes.
Grc_new_df.drop(Grc_new_df.columns[[0,6]], axis=1, inplace=True)
```

```python
# Task 6, creating new subtype DataFrames
SupType_1 = Grc_new_df[Grc_new_df['Outlet_Type'] == 'Supermarket Type1']
SupType_2 = Grc_new_df[Grc_new_df['Outlet_Type'] == 'Supermarket Type2']
```

```python
# Task 7, set style to ggplot.
plt.style.use('ggplot')
# Create subplot
figure, ax = plt.subplots(2, 1, figsize=(15, 8))
# Create boxplots
sns.boxplot(x='Item_MRP', data=SupType_1, ax=ax[0]).set_title('SupType_1')
sns.boxplot(x='Item_MRP', data=SupType_2, ax=ax[1]).set_title('SupType_2')
```

```python
# We can see that SubType_1 has a slightly higher median
```

```python
# Task 8, create Grc_Concat_df
Grc_Concat_df = pd.concat([SupType_1, SupType_2])
# Sort for Item_Outlet_Sales
Grc_Concat_df = Grc_Concat_df.sort_values('Item_Outlet_Sales')
# Display the first one, which has the lowest sales, see that it has Outlet_Location_Type Tier 3
Grc_Concat_df.head(1)
```

```python
# Task 9, set indexes
Grc_Concat_df.set_index(['Outlet_Size', 'Outlet_Location_Type'], inplace=True)
Grc_Concat_df.head(5)
```

```python
# Task 10
# Create buckets for Item_Weight
Grc_Concat_df['Weight_Buckets'] = pd.qcut(Grc_Concat_df['Item_Weight'], q=10)

# Find mean, minimum, maximum
grouped = Grc_Concat_df.groupby('Weight_Buckets')['Item_Weight']
result = grouped.agg(['mean', 'min', 'max', 'count'])

print(result)
```
